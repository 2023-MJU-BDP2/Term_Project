```python
import numpy as np
from scipy.spatial.distance import pdist, squareform

df = df_temp[df_temp['gu']=='강남구'].copy()

# haversine 함수를 벡터화
vec_haversine = np.vectorize(hs.haversine)

# 모든 점들 간의 거리 계산
distances = pdist(df[['lat', 'lng']].values, vec_haversine)

# 거리 데이터프레임 생성
dist_df = pd.DataFrame(squareform(distances), columns=df.index, index=df.index)

# 1.5보다 작은 값을 가지는 인덱스 추출
df['nears'] = dist_df[dist_df < 1.5].apply(lambda x: x.dropna().index.tolist(), axis=1)
```
