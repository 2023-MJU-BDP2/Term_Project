```python
import os
import requests
from io import BytesIO
from PIL import Image
from bs4 import BeautifulSoup

def download_image(image_url, save_directory, image_name):
    # Make a request to the image URL
    image_response = requests.get(image_url)

    # Check if the request was successful (status code 200)
    if image_response.status_code == 200:
        # Create a directory to save the image if it doesn't exist
        os.makedirs(save_directory, exist_ok=True)

        # Open the image using PIL
        image = Image.open(BytesIO(image_response.content))

        # Specify the file path to save the image
        file_path = os.path.join(save_directory, f"{image_name}.jpg")

        # Save the image to the specified file path
        image.save(file_path)

        print(f"Image downloaded successfully and saved at: {file_path}")
    else:
        print(f"Failed to download image. Status code: {image_response.status_code}")

# Example usage
search_query = "서울특별시 서대문구 통일로39길 22-25 올도우"
search_url = f"https://search.daum.net/search?w=tot&q={search_query}&DA=NTB"
soup = BeautifulSoup(search_url, "html.parser")
# You can parse the search_url to get the image URL(s) from the search results page
results = soup.find_all("img", class_="thumb_img")
# For demonstration, I'm using the same image URL you provided
image_url = "https://search1.daumcdn.net/thumb/C80x80.q80/?fname=http%3A%2F%2Ft1.daumcdn.net%2Fplace%2F471B64EBE480449D840AFD144C9ADF30"

# Call the function to download the image
download_image(image_url, "path/to/save", "example_image")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.google.com/search?q={도로명+상호명}"

try:
    response = requests.get(url)
    response.raise_for_status()  # 오류가 발생하면 예외를 일으킴

    # BeautifulSoup을 사용하여 HTML 파싱
    soup = BeautifulSoup(response.text, 'html.parser')

    # 첫 번째 <a> 태그 선택
    a_tag = soup.find('a')

    # <a> 태그의 href 속성 값 출력
    if a_tag:
        href_value = a_tag.get('href')
        print("이동하는 URL:", href_value)
    else:
        print("<a> 태그를 찾을 수 없습니다.")

except requests.exceptions.HTTPError as errh:
    print(f"HTTP 오류 발생: {errh}")
except requests.exceptions.RequestException as err:
    print(f"요청 오류 발생: {err}")
except Exception as e:
    print(f"오류 발생: {e}")
```
