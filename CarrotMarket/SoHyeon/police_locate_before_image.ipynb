```python
# '시도청'이 '서울청'인 행만 추출
seoul_police_data = POLICE_data[POLICE_data['시도청'] == '서울청']
```

```python
# 주소 값으로 위도, 경도 추출
# pip install geopy

from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
lat = []
lng =[]

for i in seoul_police_data['주소']:
    lat.append(geocoding(i)[0])
    lng.append(geocoding(i)[1])
```

```python
seoul_police_data_lotlng =  pd.DataFrame({
    'category': '경찰서',
    '주소': seoul_police_data['주소'],
    'location': seoul_police_data['경찰서'] + " " + seoul_police_data['관서명'] + seoul_police_data['구분'],
    'lat': lat,
    'lng': lng
})
```

```python
# 분류(category) : CCTV, 랜드마크, 관광명소, 가로등, 비상벨, 경찰서, 보안등, 가로등, 지하철, 버스), 이름(location, CCTV는 주소로), 위도(lat), 경도(lng), 이미지(img)

seoul_police_data_lotlng
```

```python
# 경도, 위도가 0 값인 것들 확인

zero_lat_lng = seoul_police_data_lotlng[(seoul_police_data_lotlng['lat'] == 0) & (seoul_police_data_lotlng['lng'] == 0)]
print(zero_lat_lng)
```

```python
# 카카오 API 활용 위도, 경도 추출

import requests, json
import time
 
def get_location(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    headers = {"Authorization": "KakaoAK mykey"}
    api_json = json.loads(str(requests.get(url,headers=headers).text))
    return api_json
```

```python
def result_location(index):
    api_json = get_location(zero_lat_lng.loc[index, '주소'])
    if api_json['documents']:
        address = api_json['documents'][0]['address']
        zero_lat_lng.loc[index,'lng'] = address['x']; 
        zero_lat_lng.loc[index,'lat'] = address['y']
    else:
        zero_lat_lng.loc[index,'lat'], zero_lat_lng.loc[index,'lng'] = None, None
    print(index, '번째 변환 완료...')

test_data = zero_lat_lng.copy()

for index in test_data.index:
    split=[]
    if zero_lat_lng.loc[index, '주소'].endswith(')'):
        split = zero_lat_lng.loc[index, '주소'].split('(')
        zero_lat_lng.loc[index, '주소'] = split[0]

i = 0
while i < len(zero_lat_lng.index):
    try:
        result_location(zero_lat_lng.index[i])
        i+=1
    except:
        print('time.sleep 적용합니다.')
        time.sleep(2)
        result_location(zero_lat_lng.index[i])
        i+=1
        
zero_lat_lng
```

```python
# 그럼에도 경도, 위도가 0 값인 것들 확인

after_zero_lat_lng = zero_lat_lng[(zero_lat_lng['lat'].isna()) | (zero_lat_lng['lng'].isna())]
print(after_zero_lat_lng)
```

경찰서 위도, 경도 바탕 스트리트뷰 이미지 추출

```python
import requests
from PIL import Image
import io

# 위도, 경도
lat, lon = 126.99, 37.56

# 관측 시야(Field Of View) - 최대 120 기본값 90
fov = "120"
# 방향 - 범위 0 ~ 360 (0 or 360::북, 180: 남)
heading = "-45"
# 카메라 상하 방향 설정 - 범위 -90 ~ 90 기본값 0
pitch = "30"

# 구글 API 키
google_api_key = "mykey"

# URL
url = f"https://maps.googleapis.com/maps/api/streetview?size=400x300&location={lat},{lon}&fov={fov}&heading={heading}&pitch={pitch}&key={google_api_key}"

payload = {}
headers = {}
response = requests.request("GET", url, headers=headers, data=payload)

# 이미지 바이트 데이터
bytes_data = response.content

# 이미지 변환
img = Image.open(io.BytesIO(bytes_data))
img
```
