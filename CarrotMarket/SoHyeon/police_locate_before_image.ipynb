```python
# 주소 값으로 위도, 경도 추출
# pip install geopy

from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
lat = []
lng =[]

for i in seoul_police_data['주소']:
    lat.append(geocoding(i)[0])
    lng.append(geocoding(i)[1])
```

```python
seoul_police_data_lotlng =  pd.DataFrame({
    'category': '경찰서',
    '주소': seoul_police_data['주소'],
    'location': seoul_police_data['경찰서'] + " " + seoul_police_data['관서명'] + seoul_police_data['구분'],
    'lat': lat,
    'lng': lng
})
```

```python
# 경도, 위도가 0 값인 것들 확인

zero_lat_lng = seoul_police_data_lotlng[(seoul_police_data_lotlng['lat'] == 0) & (seoul_police_data_lotlng['lng'] == 0)]
print(zero_lat_lng)
```

```python
# 카카오 API 활용 위도, 경도 추출

import requests, json
import time
 
def get_location(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    headers = {"Authorization": "KakaoAK mykey"}
    api_json = json.loads(str(requests.get(url,headers=headers).text))
    return api_json
```

```python
def result_location(index):
    api_json = get_location(zero_lat_lng.loc[index, '주소'])
    if api_json['documents']:
        address = api_json['documents'][0]['address']
        zero_lat_lng.loc[index,'lng'] = address['x']; 
        zero_lat_lng.loc[index,'lat'] = address['y']
    else:
        zero_lat_lng.loc[index,'lat'], zero_lat_lng.loc[index,'lng'] = None, None
    print(index, '번째 변환 완료...')

test_data = zero_lat_lng.copy()

for index in test_data.index:
    split=[]
    if zero_lat_lng.loc[index, '주소'].endswith(')'):
        split = zero_lat_lng.loc[index, '주소'].split('(')
        zero_lat_lng.loc[index, '주소'] = split[0]

i = 0
while i < len(zero_lat_lng.index):
    try:
        result_location(zero_lat_lng.index[i])
        i+=1
    except:
        print('time.sleep 적용합니다.')
        time.sleep(2)
        result_location(zero_lat_lng.index[i])
        i+=1
        
zero_lat_lng
```

경찰서 위도, 경도 바탕 스트리트뷰 이미지 추출

```python
import requests
from PIL import Image
import io
import csv
import os

# 구글 API 키
google_api_key = "mykey"

# CSV 파일 경로 설정
file_path = "C:/Users/user/BDP-DATA/combined_police_data_seoul.csv"

# 이미지 저장 디렉토리 경로 설정
image_directory = 'C:/Users/user/BDP-DATA/police_image'

# CSV 파일 열기
with open(file_path, 'r', encoding='utf-8') as file:
    # CSV 파일을 읽는 reader 생성
    reader = csv.reader(file)

    # 첫 번째 줄을 헤더로 읽음
    headers = next(reader)
    
     # 'lat' 컬럼과 'lng' 컬럼의 인덱스를 찾기
    lat_index = headers.index('lat')
    lng_index = headers.index('lng')
    count_index = headers.index('img')
    
     # 각 줄에 대한 작업 수행
    for row in reader:
        # 해당 줄의 'lat' 컬럼과 'lng' 컬럼 값 출력
        print(row[count_index], row[lat_index], row[lng_index])

        # URL
        url = f"https://maps.googleapis.com/maps/api/streetview?size=600x300&location={row[lat_index]},{row[lng_index]}&fov=80&heading=70&pitch=0&key={google_api_key}"
        
        response = requests.get(url)
        
        if response.status_code == 200:
            # 이미지 파일 이름 지정
            file_name = f"{row[count_index]}.jpg"
            # 이미지 파일의 저장 경로
            file_path = os.path.join(image_directory, file_name)
            with open(file_path, 'wb') as img_file:
                img_file.write(response.content)
        else:
            print("Error occurred while downloading image.")
```
